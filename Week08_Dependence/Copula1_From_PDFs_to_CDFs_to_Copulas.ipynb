```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gumbel_r, expon

# Set seed for reproducibility
np.random.seed(42)

# Generate random data for each distribution
normal_data = np.random.normal(loc=0, scale=1, size=10000)
gumbel_data = gumbel_r.rvs(loc=0, scale=1, size=10000)
exponential_data = expon.rvs(scale=1, size=10000)

# Create histogram plots for each distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, density=True, color='blue', alpha=0.7)
plt.title('Normal Distribution')

plt.subplot(1, 3, 2)
plt.hist(gumbel_data, bins=30, density=True, color='green', alpha=0.7)
plt.title('Gumbel Distribution')

plt.subplot(1, 3, 3)
plt.hist(exponential_data, bins=30, density=True, color='orange', alpha=0.7)
plt.title('Exponential Distribution')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gumbel_r, expon

# Generate data points
x = np.linspace(-3, 3, 1000)  # Adjust the range based on your data

# PDFs for each distribution
normal_pdf = norm.pdf(x, loc=0, scale=1)
gumbel_pdf = gumbel_r.pdf(x, loc=0, scale=1)
exponential_pdf = expon.pdf(x, scale=1)

# Create plots
plt.figure(figsize=(12, 3))

plt.subplot(1, 3, 1)
plt.plot(x, normal_pdf, color='blue', label='Normal')
plt.title('Normal Distribution')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, gumbel_pdf, color='green', label='Gumbel')
plt.title('Gumbel Distribution')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, exponential_pdf, color='orange', label='Exponential')
plt.title('Exponential Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gumbel_r, expon

# Generate data points
p = np.linspace(0.01, 0.99, 100)  # Adjust the range based on your data

# Pseudo-inverse (Percent-point function, quantile function) for each distribution
normal_ppf = norm.ppf(p, loc=0, scale=1)
gumbel_ppf = gumbel_r.ppf(p, loc=0, scale=1)
exponential_ppf = expon.ppf(p, scale=1)

# Create plots
plt.figure(figsize=(12, 3))

plt.subplot(1, 3, 1)
plt.plot(p, normal_ppf, color='blue', label='Normal')
plt.title('Normal Distribution (Pseudo-inverse)')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(p, gumbel_ppf, color='green', label='Gumbel')
plt.title('Gumbel Distribution (Pseudo-inverse)')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(p, exponential_ppf, color='orange', label='Exponential')
plt.title('Exponential Distribution (Pseudo-inverse)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import rankdata

# Generate random data for each distribution
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=1, size=1000)
gumbel_data = gumbel_r.rvs(loc=0, scale=1, size=1000)
exponential_data = expon.rvs(scale=1, size=1000)

# Rank data to get empirical distributions
normal_rank = rankdata(normal_data) / (len(normal_data) + 1)
gumbel_rank = rankdata(gumbel_data) / (len(gumbel_data) + 1)
exponential_rank = rankdata(exponential_data) / (len(exponential_data) + 1)

# Combine data into a 2D array for easier handling
data = np.array([normal_rank, gumbel_rank, exponential_rank]).T

# Create a DataFrame for seaborn
import pandas as pd
df_empirical = pd.DataFrame(data, columns=['Normal', 'Gumbel', 'Exponential'])

# Create a scatterplot matrix with correlation coefficients
sns.set(style="ticks")
scatter_matrix = sns.pairplot(df_empirical)

# Add correlation coefficients to the upper triangle
for i, j in zip(*np.triu_indices_from(scatter_matrix.axes, 1)):
    scatter_matrix.axes[i, j].annotate(
        f"Corr = {df_empirical.iloc[:, i].corr(df_empirical.iloc[:, j]):.2f}",
        (0.5, 0.9),
        xycoords="axes fraction",
        ha="center",
        va="center"
    )

# Show the plot
plt.show()
```
